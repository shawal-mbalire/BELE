```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

$$ KGH(s) = \frac{K}{(s+1)(s+2)(s+3)} $$

```python
# Nyquist plot
G1 = control.zpk([], [-1, -2, -3], 50) # k < 60
G2 = control.zpk([], [-1, -2, -3], 60) # k = 60
G3 = control.zpk([], [-1, -2, -3], 90) # k > 60

# creating 3 subplots
fig, axs = plt.subplots(3, 1, figsize=(5, 15))

# one subplot
plt.subplot(3, 1, 1)
control.nyquist_plot(G1)
plt.title('Nyquist plot for k < 60')

plt.subplot(3, 1, 2)
control.nyquist_plot(G2)
plt.title('Nyquist plot for k = 60')

plt.subplot(3, 1, 3)
control.nyquist_plot(G3)
plt.title('Nyquist plot for k > 60')
```

```python
plt.figure(figsize=(15, 5))

G = control.zpk([], [-1, 0, -4], 30)
plt.subplot(1, 2, 1)
control.nyquist_plot(G)
plt.title('Nyquist plot before adding pole at -25')

G = control.zpk([-25], [-1, 0, -4], 30)
plt.subplot(1, 2, 2)
control.nyquist_plot(G)
plt.title('Nyquist plot after adding pole at -25')
```
