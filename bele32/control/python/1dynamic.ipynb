```python
import numpy as np  
import matplotlib.pyplot as plt
```

```python
K = 3
T = 4

start = 0
stop  = 30
step  = 0.1

t = np.arange(start, stop, step)

y = K * (1 - np.exp(-t/T)) # simple diffential equation

plt.plot(t, y)
plt.title('First Order dynamic system')
plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```

```python
from scipy.integrate import odeint
```

```python
K = 3
T = 4
u = 1

tstart = 0
tstop  = 30
tstep  = 0.1
y0 = 0

t = np.arange(tstart, tstop, tstep)

def system1order(y, t, K, T, u):
    dydt = (1/T) * (K*u - y)
    return dydt

sol = odeint(system1order, y0, t, args=(K, T, u))# solve the ODE
plt.plot(t, sol)
plt.title('First Order dynamic system')
plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```

```python
K = 3
T = 4
a = -1/T
b = K/T

Ts = 0.1
Tstop = 30
uk = 1
yk = 0
N = int(Tstop/Ts)
data = []
data.append(yk)
for i in range(N):
    yk1 = (1 + a*Ts)*yk + b*Ts*uk
    yk = yk1
    data.append(yk)

t = np.arange(0, Tstop+Ts, Ts)

plt.plot(t, data)
plt.title('First Order dynamic system')
plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.grid()
plt.show()
```
