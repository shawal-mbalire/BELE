```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

$$ KGH = \frac{K}{s^2(s+1)} $$
$$ KGH = \frac{K}{s^2(s+1)} = \frac{K}{s^3 + s^2} $$

```python
# root locus
G = control.zpk([], [0, 0, -1], 1) # zero,pole,gain
control.root_locus(G , grid=False)
plt.grid(visible=True, which='both',axis='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)
plt.show()
```
