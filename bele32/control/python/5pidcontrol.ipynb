# PID Control

We shall use a PID controller within  afeedback loop. The PID algorithm contains three separate constant parameters, and is expressed as:

$$
u(t) = K_p e(t) + K_i \int_{0}^{t} e(\tau) d\tau + K_d \frac{de(t)}{dt}
$$

where:  
$e(t)$ is the error at time $t$,
$K_p, K_i, K_d$ are the proportional, integral, and derivative gains, respectively.


# PI Control

But for simpliticy, we shall use a PI controller, which is expressed as:

$$
u(t) = K_p e(t) + K_i \int_{0}^{t} e(\tau) d\tau
$$

where:
$e(t)$ is the error at time $t$,
$K_p, K_i$ are the proportional and integral gains, respectively.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = 3
T = 4
a = -(1/T)
b = K/T

Ts = 0.1
Tstop = 20
N = int(Tstop/Ts)
y = np.zeros(N+2)
y[0] = 0

# PI controller
Kp = 0.5
Ti = 5

r = 5
e = np.zeros(N+2)
u = np.zeros(N+2)

# Simulation
for k in range(N+1):
    e[k] = r - y[k]
    u[k] = u[k-1] + Kp*(e[k] - e[k-1]) + (Kp/Ti)*e[k]
    y[k+1] = (1 + Ts*a)*y[k] + Ts*b*u[k]

t = np.arange(0, Tstop+2*Ts, Ts)

plt.figure(1)
plt.plot(t, y)
plt.title('Control of Dynsmic system with PI controller')
plt.xlabel('Time [s]')
plt.ylabel('y')
plt.grid()
plt.show()

plt.figure(2)
plt.plot(t, u)
plt.title('Control signal u')
plt.xlabel('Time [s]')
plt.ylabel('u')
plt.grid()
plt.show()
```
