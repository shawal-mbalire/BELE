```python
import scipy.signal as sig
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = [0,0]

start = 0
stop = 30
step = 1
t = np.arange(start, stop, step)

K = 3
T = 4

A = [
    [-1/T, 0],
    [0, 0]
]
B = [
    [K/T],
    [0]
]
C = [1, 0]
D = 0

sys = sig.StateSpace(A, B, C, D)
t, y = sig.step(sys,X0=x0, T=t)

plt.plot(t, y)
plt.xlabel('Time [s]')
plt.ylabel('y')
plt.title('Step Response')
plt.grid()
plt.show()
```
